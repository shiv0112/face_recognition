```python
from mtcnn.mtcnn import MTCNN
import cv2
```

```python
cv2.destroyAllWindows()
```

```python
file_name = '../test.jpg'
 
im = cv2.imread(file_name)
cv2.imshow('Jennifer',im)
if cv2.waitKey() == 13:
    cv2.destroyAllWindows()
```

```python
detector = MTCNN()

face_detection  = detector.detect_faces(im)
```

```python
face_detection
```

```python
im_data=face_detection[0]
```

```python
im_data["keypoints"]["left_eye"]
```

```python
def bbox_plot(im, im_data):
    x,y,w,h = im_data['box']
    cv2.rectangle(im,(x,y),(x+w,y+h),(0,255,0) ,2)
```

```python
def mark_key_point(im, key):
    cv2.circle(im, (key), 1, (0,255,0), 2)
```

```python
bbox_plot(im,im_data)
mark_key_point(im, im_data['keypoints']['left_eye'])
mark_key_point(im, im_data['keypoints']['right_eye'])
mark_key_point(im, im_data['keypoints']['nose'])
mark_key_point(im, im_data['keypoints']['mouth_left'])
mark_key_point(im, im_data['keypoints']['mouth_right'])

cv2.imshow('Jennifer',im)
if cv2.waitKey() == 13:
    cv2.destroyAllWindows()
```

### MTCNN on Live Video

```python
detector = MTCNN()


cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()

    face_detection  = detector.detect_faces(frame)
    try:
        frame_data = face_detection[0]

        bbox_plot(frame,frame_data)
        mark_key_point(frame, frame_data['keypoints']['left_eye'])
        mark_key_point(frame, frame_data['keypoints']['right_eye'])
        mark_key_point(frame, frame_data['keypoints']['nose'])
        mark_key_point(frame, frame_data['keypoints']['mouth_left'])
        mark_key_point(frame, frame_data['keypoints']['mouth_right'])
        cv2.imshow("Live Video", frame)

    except:
        cv2.imshow("Live Video", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
cap.release()
```

### Multiple face

```python
import cv2
from mtcnn import MTCNN

def mark_key_point(im, key):
    cv2.circle(im, (key), 1, (0,255,0), 2)

def find_face_MTCNN(frame, result_list):
    for result in result_list:
        x, y, w, h = result['box']
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,155,255) ,2)
        mark_key_point(frame, result['keypoints']['left_eye'])
        mark_key_point(frame, result['keypoints']['right_eye'])
        mark_key_point(frame, result['keypoints']['nose'])
        mark_key_point(frame, result['keypoints']['mouth_left'])
        mark_key_point(frame, result['keypoints']['mouth_right'])
    return frame


video_capture = cv2.VideoCapture(0, cv2.CAP_DSHOW)
detector = MTCNN()

while True:
    _, frame = video_capture.read()
    faces = detector.detect_faces(frame)
    detectFaceMTCNN = find_face_MTCNN(frame, faces)
    cv2.imshow('Video', detectFaceMTCNN)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```
